```python
import seaborn as sns
import torch
from matplotlib import pyplot as plt
from torch.distributions import MultivariateNormal

from machine_learning.classes import OptimizerConfigs, TrainConfigs
from machine_learning.flows.real_nvp import RealNVP
from machine_learning.utils.utils_moons import load_moons_data
from machine_learning.utils.utils_nn import train
```

```python
plt.style.use('seaborn-whitegrid')
sns.set_color_codes()

figsize = (8, 8)
```

```python
torch.manual_seed(0)
```

```python
n_samples = 1000
n_moons_train = 5
n_moons_val = 2
n_moons_test = 1
noise_factor = 0.05

X_train_loader, X_val_loader, X_test_loader = load_moons_data(n_samples, n_moons_train, n_moons_val, 
                                                              n_moons_test, noise_factor)
```

```python
lr = 1e-2
optimizer_configs = OptimizerConfigs(lr=lr)
```

```python
n_epochs = 1000
patience = 1000
configs = TrainConfigs(optimizer_configs=optimizer_configs, n_epochs=n_epochs, patience=patience)
```

```python
x_dim = 2
h_dim = 64
z_dim = 2
n_coupling_layers = 10
```

```python
model = RealNVP(x_dim, h_dim, z_dim, n_coupling_layers=n_coupling_layers)
```

```python
best_state_dict, best_val_loss, best_epoch = train(configs, model, X_train_loader, X_val_loader, verbose=True)
```

```python
best_val_loss, best_epoch
```

```python
best_model = RealNVP(x_dim, h_dim, z_dim, n_coupling_layers=n_coupling_layers)
best_model.load_state_dict(best_state_dict)
```

### encoding moons

```python
X, _ = next(iter(X_test_loader))

Z = best_model.forward(X).detach().numpy()
plt.scatter(Z[:, 0], Z[:, 1])
plt.title(f"encoding of moons")
plt.show()
```

### generating moons

```python
Z = best_model.p.sample((1000,))
plt.scatter(Z[:, 0], Z[:, 1])
plt.title(f"Z sampled from standard MultivariateNormal")
plt.show()
```

```python
X_hat = best_model.backward(Z).detach().numpy()
plt.scatter(X_hat[:, 0], X_hat[:, 1])
plt.title(f"moons generated from Z")
plt.show()
```
