```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

from machine_learning.clustering.clusterer import k_means
```

```python
plt.style.use('seaborn-whitegrid')
sns.set_color_codes()
figsize = (12, 8)
dotsize = 30
```

### data

```python
k = 4
n_samples = 300
random_state = 0
```

```python
X, true_labels = make_blobs(n_samples=n_samples, centers=k, random_state=random_state, cluster_std=0.6)
```

```python
plt.figure(figsize=figsize)
plt.scatter(X[:, 0], X[:, 1], dotsize)
```

### scikit-learn implementation

```python
kmeans = KMeans(4)
kmeans.fit(X)
predicted_labels = kmeans.predict(X)
```

```python
plt.figure(figsize=figsize)
plt.scatter(X[:, 0], X[:, 1], dotsize, predicted_labels, cmap='rainbow')
```

### implementation uses Forgy initialization method, which depends on seed quite a bit

#### with one seed

```python
seed = 0
centers, labels = k_means(X, k, seed=seed)
```

```python
plt.figure(figsize=figsize)
plt.scatter(X[:, 0], X[:, 1], dotsize, labels, cmap='rainbow')
```

#### with another seed

```python
seed = 4  # implementation uses Forgy initialization method, which depends on seed quite a bit
centers, labels = k_means(X, k, seed=seed)
```

```python
plt.figure(figsize=figsize)
plt.scatter(X[:, 0], X[:, 1], dotsize, labels, cmap='rainbow')
```
