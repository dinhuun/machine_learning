### how 2 arrays are correlated in different scenarios

```python
import numpy as np
import os
import seaborn as sns
from matplotlib import pyplot as plt

from machine_learning.utils.utils_misc import link_relatives
```

```python
np.random.seed(0)
```

```python
figsize = (12, 8)

plt.style.use('seaborn-whitegrid')
sns.set_color_codes()
```

```python
n_samples = 100
```

#### noise

```python
noise_0 = np.random.rand(n_samples)
noise_1 = np.random.rand(n_samples)
```

```python
plt.figure(figsize=figsize)
plt.plot(noise_0, color='r')
plt.plot(noise_1, color='b')
plt.show()
```

```python
np.corrcoef(noise_0, noise_1)
```

#### noise + trend

```python
trend = np.array(list(range(n_samples)))
X_0 = noise_0 + trend
X_1 = noise_1 + trend
```

```python
plt.figure(figsize=figsize)
plt.plot(X_0, color='r')
plt.plot(X_1, color='b')
plt.show()
```

```python
np.corrcoef(X_0, X_1)
```

```python
#### difference between noise plus constant 1
```

```python
Y_0 = np.diff(X_0)
Y_1 = np.diff(X_1)
```

```python
Y_0[:10]
```

```python
plt.figure(figsize=figsize)
plt.plot(Y_0, color='r')
plt.plot(Y_1, color='b')
plt.show()
```

```python
np.corrcoef(Y_0, Y_1)
```

#### ratios of (noise + trend) which goes to limit 1

```python
Z_0 = link_relatives(X_0)
Z_1 = link_relatives(X_1)
```

```python
plt.figure(figsize=figsize)
plt.plot(Z_0, color='r')
plt.plot(Z_1, color='b')
plt.show()
```

```python
np.corrcoef(Z_0, Z_1)
```
