```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.use('seaborn-whitegrid')
sns.set_color_codes()

figsize = (12, 8)
```

```python
X = np.linspace(0, 10, 100)
Y = np.sin(X)

plt.figure(figsize=figsize)
plt.plot(X, Y)
plt.show()
```

```python
S = np.linspace(1, 10, 10)
T = np.zeros(10)

plt.figure(figsize=figsize)
plt.plot(S, T + 0.5, 'o')
plt.plot(S, T - 0.5, 'x')
plt.ylim([-1, 1])
plt.show()
```

```python
X = np.linspace(-1, 1)

fig, ax = plt.subplots(figsize=figsize)
for slope in np.arange(0, 10, 0.1):
    plt.plot(X, slope * X, '0.8')
ax.set_aspect("equal")
ax.axis([-1, 1, -1, 1])
```

```python
U = np.linspace(0, 1, 100)

plt.figure(figsize=figsize)
for alpha in range(10):
    V = 0.1 * alpha * U * U
    plt.plot(U, V)
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.show()
```

```python
n_samples = 1000
TS = pd.Series(np.random.randn(n_samples), index=pd.date_range('1/1/2000', periods=n_samples))
type(TS)
```

```python
TS.head()
```

```python
TS = TS.cumsum()
TS.head()
```

```python
TS.plot(figsize=figsize)
```

```python
n_sampels = 1000
n_features = 4
df = pd.DataFrame(np.random.randn(n_sampels, n_features), index=TS.index, columns=['A', 'B', 'C', 'D'])
df = df.cumsum()
df.head()
```

```python
df.plot(figsize=figsize)
```
