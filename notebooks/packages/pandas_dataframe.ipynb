#### categorical values

```python
m = [1, 2, 3, 4, 5, 6]
n = ['a', 'b', 'b', 'a', 'a', 'e']
data = {'id' : m, 'raw_grade' : n}
```

```python
df_4 = pd.DataFrame(data)
df_4
```

```python
df_4['grade'] = df_4['raw_grade'].astype('category')
df_4['grade']
```

```python
df_4
```

```python
df_4['grade'].cat.categories = ['very good', 'good', 'very bad']
df_4
```

#### reorder and add more grades

```python
df_4['grade'] = df_4['grade'].cat.set_categories(['very bad', 'bad', 'medium', 'good', 'very good'])
df_4['grade']
```

```python
df_4.sort_values(by='grade')
```

```python
df_4.groupby('grade').size()
```
